# Find data on days going back to ComCam on sky

```python
from lsst.daf.butler import Butler
from lsst.daf.butler.registry import Registry
```

```python
repo = 'main'
instrument = 'LSSTComCam'
datasetType = 'raw'
# raw all collection contains all images 
collections = ["LSSTComCam/defaults"]

butler = Butler(repo, collections=collections, instrument=instrument)

registry = butler.registry

#uncomment to get the help page
#help(registry) 
```

```python
day = 20241207
dims = ['exposure']
result = registry.queryDataIds(  datasets=datasetType, dimensions = dims,
          where = f"day_obs = {day} "
                              )
results = [r for r in result]

print (f"got {len(results)} images")
```

```python
day = 20241207
dims = ['exposure']
result = registry.queryDataIds(  datasets=datasetType, dimensions = dims,
          where = f"day_obs = {day} "
                              )
results = [r for r in result]

print (f"got {len(results)} images")
exp=""
exposures = []
for r in results:
   if r['exposure'] != exp:
        exp=  r['exposure']
        exposures.append(exp)

print (f"there are {len(exposures)} exposures")
datasetRefs = registry.queryDatasets(datasetType=datasetType,
                                     where=f'exposure = {exposures[3]}')

#get one image just to show i t exists
results = [r for r in datasetRefs]

exp = results[1]
img = butler.get(datasetType,exp.dataId)
assert (img is not None)
```

This is just getting the datasetref  the first exposure 

```python
def checkDay (registry,day, datasetType) -> int:
    dims = ['exposure']
    result = registry.queryDataIds(  datasets=datasetType, dimensions = dims,
              where = f"day_obs = {day} "
                                  )
    results = [r for r in result]
    
    exposure = results[1]['exposure']
    datasetRefs = registry.queryDatasets(datasetType=datasetType,
                                         where=f'exposure = {exposure}')
    #get one image just to show it exists
    results = [r for r in datasetRefs]
    
    exp = results[1]
    img = butler.get(datasetType,exp.dataId)
    assert (img is not None)
    return len(results)
```

# Check some ComCam

```python
repo = 'main'
instrument = 'LSSTComCam'
datasetType = 'raw'
# raw all collection contains all images 
collections = ["LSSTComCam/defaults"]
butler = Butler(repo, collections=collections, instrument=instrument)
registry = butler.registry

#ComCam
for day in [20241127, 20241207, 20241208]:
   count = checkDay(registry, day = day, datasetType = datasetType)
   print (day, count)
```
